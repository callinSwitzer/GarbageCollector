```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.xkcd()
plt.rcParams.update({'font.size': 36, 'axes.linewidth': 4})

fig = plt.figure(figsize = (10,7))
ax = fig.add_subplot(1, 1, 1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.xticks([])
plt.yticks([])

data = np.ones(100)
data[20:] += np.arange(80)**2
ax.set_ylim([-1000, np.max(data) + 1000])


# plt.annotate(
#     'THE DAY I REALIZED\nI WOULN\'T FINISH \nMY PHD AT THIS RATE',
#     xy=(50, 9.620e+02), arrowprops=dict(arrowstyle='->', linewidth = 4), xytext=(2, 3000), 
# )

plt.plot(data, linewidth= 4)


plt.xlabel('time')
plt.ylabel('Data size')

plt.tight_layout()
plt.savefig(r"D:\Dropbox\CallinStuff\MSDSE_2018_ParkCity\dataSizeplot.svg")
```
