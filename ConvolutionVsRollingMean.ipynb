Is a convolution the same as a rolling mean?




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.ion()

import sys

print(sys.version)
print(sys.exec_prefix)
```

```python
exData = np.random.randn(100)
```

```python
plt.plot(exData)
```

```python
rollMean = pd.Series(exData).rolling( window= 10, center = True).mean()
```

```python
plt.plot(rollMean)
```

```python
# 1D convolution
def conv_movingAvg(x, w):
    '''
    x = data
    w = window length
    '''
    k = np.ones(w) / w
    y = np.convolve(x, k, 'same')
    return(y)
```

```python
conv1 = conv_movingAvg(exData, 10)
```

```python
plt.plot(conv1, label = "pd rolling mean")
plt.plot(rollMean, label  = 'np convolve')
plt.legend()
```
