```python
import cv2
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
plt.ion()

import pandas as pd
```

```python
# read in image
img = cv2.imread(r"C:\Users\calli\Desktop\6.16_IF1_LF1_crop.jpeg")

# reshape
im_res = np.reshape(img, (-1, 3))

# remove (255, 255, 255), which is white
im_res_del = np.delete(im_res, np.where(im_res >= [255, 255, 255]), axis=0)

print(np.mean(im_res_del, axis = 0))

io.imshow(img)
plt.show()
plt.hist(im_res_del[:,1], bins = 50, label = 'green', color = 'g', alpha = 0.5)
plt.hist(im_res_del[:,0], bins = 50, label = 'blue', color = 'b',  alpha = 0.5)
plt.hist(im_res_del[:,2], bins = 50, label = 'red', color = 'r',  alpha = 0.5)
plt.xlabel("pixel values")
plt.ylabel("count")
plt.legend()
plt.show()
```

```python
# Standard imports
 
# Read image
im = cv2.imread(r"C:\Users\calli\Desktop\6.16_IF1_LF1.jpeg", cv2.IMREAD_GRAYSCALE)
 
# Set up the detector with default parameters.
detector = cv2.SimpleBlobDetector_create()
 
# Detect blobs.
keypoints = detector.detect(im)
 
# Draw detected blobs as red circles.
# cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS ensures the size of the circle corresponds to the size of blob
im_with_keypoints = cv2.drawKeypoints(im, keypoints, np.array([]), (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
 
# Show keypoints
plt.figure(figsize = [10,10])
io.imshow( im_with_keypoints)
```
