```python
# read in videos and combine them frame by frame

# reset frame num to zero
cap.set(cv2.CAP_PROP_FRAME_COUNT, 0) 
cap2.set(cv2.CAP_PROP_FRAME_COUNT, 0) 

ret = True
ctr = 1
ctr2 = 1

stt = time.time()
while(ctr < length):  
    # capture frame-by-frame
    ret, frame = cap.read()
    ret2, frame2 = cap2.read()
    if not ret:
        break
    if((ctr > 198) and (ctr < 500)):
        if np.mod(ctr,1) == 0:
            # process frames

            #crop
            frame = frame[205:, 105:610]
            frame2 = frame2[205:, 105:610]

            frame = frame / np.max(frame)
            frame2 = (frame2 - np.min(frame2) )/ np.max(frame2)

            kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(2,2))
            dilation = cv2.dilate(frame2,kernel,iterations = 1)
            erosion = cv2.erode(dilation,kernel,iterations = 1)


            ff= frame.copy()
            ff[np.round(erosion) > 0 ] = 0

            frame = frame*255.0
            ff = ff*255.0



            # combine frames
            combFrames = np.concatenate([frame, ff], axis = 0)

            cv2.imwrite(str(ctr2).zfill(4) + ".png", combFrames)
            ctr2+=1
    print(ctr)
    ctr += 1

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
# When everything is done release the capture
cap.release()
cap2.release()
cv2.destroyAllWindows()

print(time.time() - stt) # 
```
