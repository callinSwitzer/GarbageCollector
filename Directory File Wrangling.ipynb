```python
# remove videos that aren't listed as outdoor videos
vidKeep = pd.read_csv("/Users/cswitzer/Dropbox/ExperSummer2016/Kalmia/KalmiaOutdoorVids/001_KalmiaInsectVisitors_OutdoorONLY.csv")
```

```python
moviesKeep = [f + ".avi" for f in vidKeep.Video]
```

```python
moviesDelete = set(movies) - set(moviesKeep) 
```

```python
moviesDelete
```

```python
#[re.sub(".avi", "", mm) for mm in moviesDelete]
namesToDelete = [re.sub(".avi", "", mm) for mm in moviesDelete]
```

```python
for ii in os.listdir(direct):
    starter = [ii.startswith(jj) for jj in namesToDelete]
    if sum(starter) > 0:
        os.remove(os.path.join(direct, ii))
```
